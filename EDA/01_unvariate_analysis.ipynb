```python
import pandas as pd
```

```python
df = pd.read_csv("01_train.csv")
df.sample(5)
```

This is titanic dataset. 
- Survived defines if the passenger survived(1) or not(0).
- Pclass is passenger class with 1 being most premium and then so on.
- SibSp -> if a sibling or spouse is traveling with that person.
- Parch -> if a parent or child is traveling with that person.
- Fare -> the fare paid by the passenger for the ticket.
- Embarked -> where did they embark (3 stations)

# 1. Categorical columns 
- survived 
- Pclass
- sex
- Sibsp
- Parch
- Embarked

For categorical columns we can do these things:
- find out the count of each type.
- plot the countplot/bar graph 
- plot the pie chart(with percentage distribution)

## a. countplot

```python
import seaborn as sns
```

```python
df['Survived'].value_counts()
```

```python
sns.countplot(x='Survived',data=df)
```

```python
sns.countplot(x='Pclass',data=df)
```

```python
df['Pclass'].value_counts()
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

most people are in class 3 and then class 1 then class 2

```python
sns.countplot(x='Sex',data=df)
```

there are more males than females

```python
sns.countplot(x='Embarked',data=df)
```

most people boarded from S then C then Q

## b. piechart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f') # using autopct we can see the % of value counts
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct="%.2f")
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct="%.2f")
```

# 2. Numerical Columns

In numerical columns we generally perform these actions:
- draw *histogram* to see the distribution of data(bins can be defined)
- draw *distplot/kdeplot*  to see the probability distribution function of the data
- draw *boxplot* to get 5 number summary 

## a. histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'])
```

we can see the age distribution here. Most people are in 20s-40s

## b.distplot

```python
sns.displot(x='Age',data=df,kde=True)
```

we are plotting the probability distribution here 

## c. boxplot

```python
sns.boxplot(x='Age',data=df)
```

```python
sns.boxplot(x='Fare',data=df)
```

we use boxplot to find out the min,max,median and the outliers

we can get the skewness of data too.
- Symmetric distribution (≈0): Balanced spread of data around the mean.
- Positively skewed/left skewed (>0): Indicates outliers or longer tail on the higher side.
- Negatively skewed/ right skewed (<0): Indicates outliers or longer tail on the lower side.

```python
df['Age'].skew()
```

```python
sns.kdeplot(data=df['Age'],fill=True)
```

left skewed as can be seen in graph

```python
df['Fare'].skew()
```

```python
sns.kdeplot(data=df['Fare'])
```

left skewed
